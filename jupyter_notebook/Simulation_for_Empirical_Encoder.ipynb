```python
import torch
from torchvision import datasets, transforms
import numpy as np
import matplotlib.pyplot as plt

# Load MNIST dataset
transform = transforms.Compose([transforms.ToTensor()])
mnist_data = datasets.MNIST(root='./data', train=True, download=True, transform=transform)

# Function to generate y values
def generate_y(t, is_even):
    if is_even:
        return np.random.normal(0, 0.1 + 2 * (np.cos(np.pi * t) ** 2))
    else:
        return np.random.normal(0, 0.1 + 2 * (np.sin(np.pi * t) ** 2))

# Simulation parameters
n_samples = 1000

# Arrays to store results
positions = []
y_values = []
digits = []

# Run simulation
for _ in range(n_samples):
    # Get random MNIST image
    idx = np.random.randint(0, len(mnist_data))
    image, label = mnist_data[idx]
    
    # Generate random position
    t = np.random.uniform(-0.5, 2.5)
    
    # Determine if digit is even or odd
    is_even = label % 2 == 0
    
    # Generate y value
    y = generate_y(t, is_even)
    
    # Store results
    positions.append(t)
    y_values.append(y)
    digits.append(label)

# Convert to numpy arrays
positions = np.array(positions)
y_values = np.array(y_values)
digits = np.array(digits)

# Plotting
plt.figure(figsize=(12, 8))
plt.scatter(positions[digits % 2 == 0], y_values[digits % 2 == 0], 
            alpha=0.5, label='Even digits')
plt.scatter(positions[digits % 2 != 0], y_values[digits % 2 != 0], 
            alpha=0.5, label='Odd digits')
plt.xlabel('Position (t)')
plt.ylabel('Y value')
plt.title('Simulation Results')
plt.legend()
plt.grid(True)
plt.show()

# Print some statistics
print(f"Number of samples: {n_samples}")
print(f"Number of even digits: {np.sum(digits % 2 == 0)}")
print(f"Number of odd digits: {np.sum(digits % 2 != 0)}")
print(f"Mean y value for even digits: {np.mean(y_values[digits % 2 == 0]):.4f}")
print(f"Mean y value for odd digits: {np.mean(y_values[digits % 2 != 0]):.4f}")
print(f"Std dev of y for even digits: {np.std(y_values[digits % 2 == 0]):.4f}")
print(f"Std dev of y for odd digits: {np.std(y_values[digits % 2 != 0]):.4f}")

# Histogram of y values
plt.figure(figsize=(12, 6))
plt.hist(y_values[digits % 2 == 0], bins=50, alpha=0.5, label='Even digits')
plt.hist(y_values[digits % 2 != 0], bins=50, alpha=0.5, label='Odd digits')
plt.xlabel('Y value')
plt.ylabel('Frequency')
plt.title('Distribution of Y values')
plt.legend()
plt.grid(True)
plt.show()
```
